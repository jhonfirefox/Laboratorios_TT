```python
#Este programa permite graficas los datos de reflectancia en funcion
# de la longitud de onda, donde se le ha aplicado un filtro savgol

import numpy as np
import matplotlib.pyplot as plt
import random as rand
import pandas as pd
```

```python
url  = "https://raw.githubusercontent.com/jhonfirefox/Laboratorios_TT/main/Lab1_DA.csv"

Dlist1 = pd.read_csv(url, sep=",", decimal=".", encoding='windows-1252')
Dlist1
```

```python
#Se eliminan las ultimos dos registros que no aporta nada a los datos
Dlist1 = Dlist1.drop(17)
Dlist1 = Dlist1.drop(18)
Dlist1
```

```python
#Se convierte a un valor numerico la id de cada paciente
Dlist1["id"] = Dlist1["id"].astype(int)
Dlist1
```

```python
#Contamos los valores unicos la columna pais
#Colocamos en letra minuscula los paises
Dlist1["pais"] = Dlist1["pais"].str.lower()
Dlist1["pais"].value_counts()
```

```python
#Primero contamos el numero de paises en el data set
numeroPaises = Dlist1["pais"].value_counts().index.tolist()
print(numeroPaises)
```

```python
#Se estandariza el valor del pais

#Los indices 1, 4, 5, 6, 7, 8, 9 representan registro de pais incorrecto

paisCorrecto = ["colombia", "chile", "japon", "japon", "peru"]
pais = ["colombiaaa", "chiles", "japonn", "rapon", "pero"]

for i, paisT in enumerate(pais):
  Dlist1.replace(to_replace = paisT, value = paisCorrecto[i], inplace = True)

Dlist1
```

```python
#Primero contamos el numero de servicios en el data set
numeroServicio = Dlist1["servicio"].value_counts().index.tolist()
print(numeroServicio)
```

```python
servicioCorrecto = ["hogar", "hogar", "comercio", "comercio", "comercio", "loguear"]
servicio = ["hogarr", "rogar", "comerciio", "comerciooo", "comerrr", "loguearrr"]

for i, servicioT in enumerate(servicio):
  Dlist1.replace(to_replace = servicioT, value = servicioCorrecto[i], inplace = True)

Dlist1
```

```python
#Verificamos los duplicados
Dlist1['nombres y apellidos'].duplicated()
```

```python
#Identificamos los indices duplicados
index1 = Dlist1[Dlist1['nombres y apellidos'].duplicated()].index.to_list()

for i in index1:
  Dlist1.drop(i, inplace = True)

Dlist1
```

```python
#Eliminamos los valores NaN de las columna valor de factura

Dlist1 = Dlist1.dropna(subset=['valor factura'])
Dlist1
```
