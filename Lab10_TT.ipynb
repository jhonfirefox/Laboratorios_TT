```python
#Se importan las librerias necesarias para los codigos
import pandas as pd
import numpy as np
from cryptography.fernet import Fernet
import datetime
```

```python
#Definimos algunas funciones que se utilizaran con frecuencia
def validacionDatos_Numerico(datos, campo):
  datos['Validacion_'+campo] = datos[campo].apply(lambda x:'Válido' if (pd.notnull(x) and (isinstance(x, int) or isinstance(x, float))) else 'Inválido')

def convertir_a_entero(valor):
    try:
        return int(valor)
    except ValueError:
        return valor

#Funcion para validar que los datos sean no negativos
def validacionDatos_Numerico_NN(datos, campo):
  datos['Validacion_NN_'+campo] = datos[campo].apply(lambda x:'Válido' if (pd.notnull(x) and (isinstance(x, int) or isinstance(x, float)) and x >= 0) else 'Inválido')

def valoresFaltantes(datos):
    return datos.isnull().sum()

##Corregimos la fecha de los registros que estan en formato "%Y/%m/%d"
def corregir_fecha(fecha_str, formato):
    try:
        fecha = datetime.datetime.strptime(fecha_str, formato)
        return fecha.strftime("%Y-%m-%d")
    except ValueError:
        return fecha_str

def validarFecha(fecha_str, formato):
    try:
        fecha = datetime.datetime.strptime(fecha_str, formato)
        return "Válido"
    except ValueError:
        return "Inválido"
def eliminarFechasNulas(datos, campo):
  for indice, fecha in enumerate(datos[campo]):
    if pd.isnull(fecha):
      datos = datos.drop(indice)
  return datos

#Validamos que el campo de la fecha se encuentre correcto
def validacionDatos_Fecha(datos, campo, formatoFecha):
   #datos['Validacion'+campo] = datos[campo].apply(lambda x:'Válido' if isinstance(datetime.datetime.strptime(x, "%Y/%m/%d"), datetime.datetime) else 'Inválido')
   datos['Validacion_'+campo] = datos[campo].apply(validarFecha, args=(formatoFecha,))
```

```python
#Generamos las listas con los datos
datos3 = ["Juan,25,juan@mail.com,colombia,M,5",
"Maria,30,maria@mail.com,Colombia,F,7",
"Pedro,,pedro@mail.com,Mexico,H,6",
"Ana,29,ana@mail.com,Argentina,F,4",
"Luisa,28,luisa@mail.com,Colombia,Femenino,8",
"Carlos,200,carlos@mail.com,Brazil,M,2",
"Juan,25,juan@mail.com,colombia,M,5"]

Nombre =[]
Edad = []
Email = []
Pais = []
Genero = []
Visitas = []

listaT = [Nombre, Edad, Email, Pais, Genero, Visitas]

#Parte del codigo para separar los campos
for i in range(len(datos3)):
  #Seleccionamos cada linea y partimos las cadadena cada ","
  datos = datos3[i].split(",")

  for j in range(len(datos)):
    listaT[j].append(datos[j])

for i in range(len(listaT)):
  print(listaT[i])
```

```python
#Generamos las listas con los datos

datos3_dic = {"Nombre": Nombre, "Edad": Edad, "Email": Email, "Pais": Pais, "Genero": Genero, "Visitas": Visitas}
#print(datos1)

for key in datos3_dic.keys():
    print(key, ":", datos3_dic[key])
```

```python
#convertimos el diccionario a un dataframe
datos3_df = pd.DataFrame(datos3_dic)
datos3_df
```

```python
#Eliminamos valores duplicados
datos3_df = datos3_df.drop_duplicates()
datos3_df
```

```python
#Verificamos el no haya valores faltantes en las columnas
valoresFaltantes(datos3_df)
```

```python
#Para el caso de la columna edad verificamos que los datos sean numericos
validacionDatos_Numerico(datos3_df, 'Edad')
datos3_df
```

```python
#Convertimos a entero y validamos la edad
datos3_df['Edad'] = datos3_df['Edad'].apply(convertir_a_entero)

#validamos la edad
validacionDatos_Numerico(datos3_df, 'Edad')
datos3_df
```

```python
#Eliminamos datos atipicos edades que sean negativas o mayores que un valor
maximaEdad = 50

for index in datos3_df.index:
  if datos3_df.loc[index, 'Validacion_Edad'] == "Válido":
    if (datos3_df.loc[index, 'Edad'] < 0 or datos3_df.loc[index, 'Edad'] > maximaEdad):
      datos3_df = datos3_df.drop(index)

datos3_df
```

```python
#Reemplazar los valores NaN o vacio con un valor específico(media, mediana)
mediaEdad = datos3_df.loc[datos3_df['Validacion_Edad']== 'Válido', 'Edad'].mean()
#Se debe convertir a entero
mediaEdad = int(mediaEdad)

for index in datos3_df.index.values:
  if (datos3_df.loc[index, 'Validacion_Edad'] == "Inválido"):
    datos3_df.loc[index, 'Edad'] = mediaEdad

datos3_df
```

```python
#validamos la edad
validacionDatos_Numerico(datos3_df, 'Edad')
datos3_df
```

```python
#Estandarizamos los valores de pais
datos3_df['Pais'] = datos3_df['Pais'].str.lower()
datos3_df
```

```python
#Solo debe haber M: Masculino y F:Femenino
boolM = datos3_df['Genero'] != 'M'
boolF = datos3_df['Genero'] != 'F'
boolM = boolM.to_list()
boolF = boolF.to_list()

print(boolM)
print(boolF)
print(np.array(boolM) & np.array(boolF))
#datos3_df.loc[np.array(boolM) & np.array(boolF), 'Genero']

#guardamos los incides de los valores distintos
valuesGenero = datos3_df.loc[np.array(boolM) & np.array(boolF), 'Genero'].values
indexGenero = datos3_df.loc[np.array(boolM) & np.array(boolF), 'Genero'].index.values
print("Indices de genero a cambiar", indexGenero)
print("Valores de genero a cambiar", valuesGenero)
#
#cambiamos los valores
for i in range(len(indexGenero)):
  if valuesGenero[i] == 'H':
    datos3_df.loc[indexGenero[i], 'Genero'] = 'M'
  if valuesGenero[i] == 'Femenino':
    datos3_df.loc[indexGenero[i], 'Genero'] = 'F'

datos3_df
```
